For this notebook, I'll be using NewsApI to extract News articles from CNN.

```python
import requests 
import config as cfg
import json
import pandas as pd
```

Create a function to extract articles from CNN. I used a 'page' parameter for the user to flip through different pages and load more articles within the parameters.

```python
def NewsFromCNN(page):
     
    # following query parameters are used
    # sources, sortBy and apiKey, from, to, page
    query_params = {
      "sources": "cnn",
      "sortBy": "popularity",
      "apiKey": cfg.api["apiKey"],
        "from": "2021-03-11",
        "to": "2021-03-12",
        "page": page
    }
    main_url = " https://newsapi.org/v2/everything"
 
    # fetching data in json format
    res = requests.get(main_url, params=query_params)
    open_cnn_page = res.json()
 
    # getting all articles in a string article
    article = open_cnn_page["articles"]
    
 
    # empty list which will 
    # contain all trending news
    results = []
     
    for ar in article:
        results.append({
            "date": ar["publishedAt"],
            "title": ar["title"],
            "author": ar["author"],
            "brief_article": ar["content"],
            "article_link": ar["url"]
        })
    return results
```

Created a file writing function to write the given data into a JSON file

```python
def writeToJSONFile(path, fileName, data):
    filePathNameWExt = './'+ path + '/' + fileName + '.json'
    with open(filePathNameWExt, 'w') as fp:
        json.dump(data, fp)
```
